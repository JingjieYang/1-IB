```python
import matplotlib.pyplot as plt
import numpy as np

# mean monthly maximum temperatures (ºC) for Christchurch, New Zealand
xs = np.arange(12) + 1  # months
ys = np.array([21, 20, 18.5, 15, 11.5, 9, 8, 9, 11, 14.5, 17, 19])  # temperature (ºC)
plt.scatter(xs, ys)
plt.show()
```

```python
def create_model(a, b, c, d):
    """y = A sin(B(x - c)) + D"""
    return lambda X: a * np.sin(b * (X - c)) + d

def calculate_error(xs, theoretical, experimental):
    return np.sum((theoretical - experimental) ** 2)
```

```python
# Model 1
model_jingjie = create_model(6, np.pi / 6, 10, 14.75)
calculate_error(xs, ys, model_jingjie(xs))
```

```python
# Model 2
model_edern = create_model(6.5, np.pi / 6, 10, 14.5)
calculate_error(xs, ys, model_edern(xs))
```

```python
# Edern wins!
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

as_ = np.arange(5, 7.5, 0.1)
ds = np.arange(13, 16, 0.1)
X, Y = np.meshgrid(as_, ds)
zs = np.array([calculate_error(xs, ys, create_model(a, np.pi / 6, 10, d)(xs)) for a, d in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('vertical dilation')
ax.set_ylabel('vertical shift')
ax.set_zlabel('squared error');
```
