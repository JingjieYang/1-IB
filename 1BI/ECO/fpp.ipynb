# La frontière des possibilittés de production (fpp)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


def plot_curves(title, xlabel, ylabel, *curve_intercepts):
    plt.figure()
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    LINE_STYLES = ['-', '-.', '--', ':']
    LINE_COLORS = ['k', 'r', 'g', 'b']
    for i, (x_intercept, y_intercept) in enumerate(curve_intercepts):
        xs = np.linspace(0, x_intercept)
        ys = y_intercept * np.sqrt(1 - (xs / x_intercept) ** 2)
        plt.plot(xs, ys, LINE_STYLES[i] + LINE_COLORS[i])
    plt.xlim([0, 5])
    plt.ylim([0, 3])
    plt.xticks([])
    plt.yticks([])
```

## production ~ richesse

### 1. Fpp dans les pays à haut revenu

```python
plot_curves("Fpp dans les pays à haut revenu", "Bien de luxe", "Bien de base", (1, 2), (3, 2.5))
```

### 2. Fpp dans les pays à faible revenu

```python
plot_curves("Fpp dans les pays à faible revenu", "Bien de luxe", "Bien de base", (1, 2))
```

### 3. Economie moderne

```python
plot_curves("Economie moderne", "Bien privé", "Bien publique", (3, 1), (3.25, 2.25))
```

### 4. Economie pauvre

```python
plot_curves("Economie pauvre", "Bien privée", "Bien publique", (3, 1))
```

## d'autres représentations

### 5. Fpp présente et futur

```python
plot_curves(
    "fpp présente et futur", "Consommation actuelle", "Investissement en capital",
    (1.25, 1), (2.25, 2), (2.5, 2.25), (2.75, 2.5)
)
```

Cette tendance n'est pas infinite:  
une augmentation d'investissement ne correspond pas forcément à la croissance de la consmomation.

### 6. Fpp de plusieurs pays

```python
plot_curves(
    "fpp des plusieurs pays",
    "Consommation courante",
    "I. en capital",
    (0.75, 0.5),
    (1.25, 1),
    (1.75, 1.5),
    (2.25, 2)
)
plt.legend(["base", "pays 1", "pays 2", "pays 3"]);
```

Supposons 3 pays qui partent de la même situation de base représentée par la fpp la plus à gauche à la figure.  
Les 3 pays sont en différents neveaux d'investissement:

- le pays 1 investit peu, il se contente de remplacer les machines;
- le pays 2 investit davantage, ce qui lui permet de consommer en avantage;
- le pays 3 fait des investissements de productivité.

Plus les pays investissent, plus ils ont des consommations futures importantes.

### 7. Fpp en économie internationale

```python
plot_curves("fpp en économie internationale", "Pétrole", "Tuyau", (3, 2), (4, 2.1))
plt.legend(["fpp 1", "fpp 2 (l'augmentation de l'exportation)"]);
```
