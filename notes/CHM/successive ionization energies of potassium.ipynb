```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = np.arange(1, 20)
IE = np.array([419, 3051, 4412, 5877, 7975, 9649, 11343, 14942, 16964, 48577,
               54433, 60701, 68896, 75950, 83152, 93403, 99771, 444911, 476075
              ])
IE_log10 = np.log10(IE)
```

```python
plt.plot(n, IE_log10, 'x-')
plt.title("Successive Ionization Energies of Potassium")
plt.xticks(n)
plt.xlabel("nth Ionization Energy")
plt.ylabel("log10(IE) $(kJ mol^{-1})$")
plt.show()
```

The ionization energy increases progressively when removing electrons in the same shell, due to the increasing effective nuclear charge; the increment of the ionization energy is significant when we start to remove electrons from a lower energy shell.

$Z_{eff}$: effective nuclear charge

$Z_{eff} = \text{charge of the nucleus} - \# \text{core } e^-s$

```python
def determine_z_eff_of_k_ions(electrons_removed_count):
    charge_nucleus = 18 + electrons_removed_count
    electrons_remaining_count = 19 - electrons_removed_count    
    if electrons_remaining_count <= 2:
        core_electrons_count = 0
    elif electrons_remaining_count <= 10:
        core_electrons_count = 2
    elif electrons_remaining_count <= 18:
        core_electrons_count = 10
    else:
        core_electrons_count = 18
    return charge_nucleus - core_electrons_count
```

```python
z_effs = [determine_z_eff_of_k_ions(i - 1) for i in n]
```

```python
plt.figure()
plt.plot(n - 1, z_effs, 'o-r')
plt.xticks(n - 1)
plt.xlabel("number of electrons removed")
plt.ylabel("effective nuclear charge")
plt.show()
```
