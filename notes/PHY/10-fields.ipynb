## Electric field strength
$\begin{aligned}
\vec{E} &= \frac{\vec{F}}{q} \\
\frac{[N]}{[C]} &= \frac{[N][m]}{[C][m]} = \frac{[J]}{[C][m]} = \frac{[V]}{[m]}
\end{aligned}$

Electric field strength is the force acting on the **positive** test charge per test charge.

$\begin{aligned}\vec{F} = \frac{k q Q}{\vec{r}^2} \Rightarrow \vec{E} = \frac{k Q}{\vec{r}^2}\end{aligned}$ 
where Q is the charge of the field

*Electric field strength at distance $r$ from $Q$*

## Gravitational field strength
$\begin{aligned}
\vec{g} &= \frac{\vec{F}}{m} \\
\frac{[N]}{[kg]} &= \frac{[kg][m][s]^{-2}}{[kg]} = \frac{[m]}{[s]^2} \\
\vec{W} &= m \vec{g} \\
\vec{F} &= m \vec{g}
\end{aligned}$

Gravitational field strength is the force on a small test charge of mass  / mass test charge m.

$\begin{aligned}
\vec{F} = \frac{G m M}{\vec{r}^2} \Rightarrow \vec{g} = \frac{GM}{\vec{r}^2}
\end{aligned}$


```python
# Earth
G = 6.67e-11  # gravitational constant
M = 6e24      # mass of Earth
r = 6400_000  # equator
g = G * M / r**2
print(g)
```

```python
# Moon
from math import sqrt

G = 6.67e-11  # gravitational constant
M = 7.4e22    # mass of moon
g = 1.6       # acceleration of free fall on moon
r = sqrt(G * M / g ** 2)
print(r)
```

## Conservative forces
- **conservative force**: any force which does work independent of path
- gravitational force is a conservative force
- friction is not conservative
- electrostatic force is a conservative force

Conservative forces have a potential function attached to them:

$\begin{aligned}
G_{P E} &= E_p = mgh\\
E_{P E} &= qV
\end{aligned}$

## Equations for fields
### Potential energy
$\begin{aligned}
\overbrace{V_e}^{\text{Electric fields}} &= \frac{k Q}{r}\\
\overbrace{V_g}^{\text{Gravitational fields}} &= \frac{G m}{r}
\end{aligned}$

### Field strength
$\begin{aligned}
&= -\underbrace{\frac{\Delta V_e}{\Delta r}}_{\text{Potential gradient}}\\
g &= -\overbrace{\frac{\Delta V_g}{\Delta r}}^{\text{Potential gradient}}
\end{aligned}$

## Escape velocity
The velocity at which you need to launch a mass $m$ from the surface of mass $M$ for it to escape gravitational forces.

- using energy considerations, find an expression for $v$ in terms of $G$, $M$, $r$.

$\begin{aligned}
E_{T \ initial} &= \overbrace{E_{p \ initial}}^{-} + \overbrace{E_{k \ initial}}^{+}\\
E_{T \ final}   &= \underbrace{E_{p \ final }}_{0} + \underbrace{E_{k \ final }}_{0}\\
E_{T} &= -m \frac{G M}{r} + \frac{1}{2} m v^2 = 0\\
\frac{1}{2} m v^2 &= \frac{G M m}{r}\\
v^2 &= \frac{2 G M}{r}\\
\Rightarrow v &= \sqrt{\frac{2 G M}{r}}\\
v &= \sqrt{2 g r}
\end{aligned}$

## Some concepts of circular motion
- Motion like this: 〇  
- $T$ is the period of revolution (s) $= \frac{1}{f}$  
- $f$ is the frequency of revolution (Hz) $= \frac{1}{T}$  
- Angular speed $\omega = \frac{\theta}{t}$ (rad / s)
- $T = \frac{2 \pi}{\omega}$
- $\omega = 2 \pi f$
- $a = \frac{v^2}{r}$
- $\vec{F} = m \vec{a}$

### a) Find an expression for $\omega^2$ in terms of $G, M_e, r^3$
$\begin{aligned}
T &= \frac{2 \pi r}{v}\\
\omega &= \frac{2 \pi}{T} = \frac{2 \pi}{\frac{2 \pi}{v}} = \frac{v}{r}\\
\omega^2 &= \frac{v^2}{r^2}\\
F &= m a\\
\frac{G M_e m}{r^2} &= m \frac{v^2}{r}\\
\frac{v^2}{r^2} &= \frac{G M_e}{r^3}\\
\omega^2 &= \frac{G M_e}{r^3}
\end{aligned}$

### b) Hence, find an expression for $T^2$ as a function of $G, M_e, r^3$
$\begin{aligned}
T &= \frac{2 \pi}{\omega}\\
T^2 &= \frac{4 \pi^2}{\omega^2}\\
&= \frac{4 \pi^2}{\frac{G M_e}{r^3}}\\
&= \frac{4 \pi^2 r^3}{G M_e}
\end{aligned}$
